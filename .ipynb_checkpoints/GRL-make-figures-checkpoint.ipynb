## This notebook reproduces Figure 2 and Figure 3 in the main text, and Supplementary Information Figure S2. 

### Functions needed to make the plots:

```python
def cumul_prob(data):
    data = np.sort(data)
    return np.array([(data[i] < data).sum() for i in range(len(data))])

def normed_CPD(data):
    data = np.sort(data)
    prob = [(data[i] < data).sum() for i in range(len(data))]
    return prob / max(prob)
```

### Prepare Figure 2

### Import the data:

```python
data = pd.read_csv('Ngozumpa-data/2019_Ngozumpa-depressions-ponds-master.csv')
print("Column Headers: ", list(allData))
```

### Create the figure

```python
data_withPond = data[data['ponded'] == 'yes']
data_withoutPond = data[data['ponded'] == 'no']

areas = data['area']
perimeters = data['perimeter']
depths = data['depth']

areas_with = data_withPond['area']
perimeters_with = data_withPond['perimeter']
depths_with = data_withPond['depth']

areas_without = data_withoutPond['area']
perimeters_without = data_withoutPond['perimeter']
depths_without = data_withoutPond['depth']

#colorblind-friendly palette
#https://yoshke.org/blog/colorblind-friendly-diagrams
c1 = '#E69F00' 
c2 = '#56B4E9' 
c3 = '#F0E442' 
c4 = '#009E73' 
c5 = '#D55E00' 
c6 = '#0072B2'

fig, ax = plt.subplots(3,1, figsize = (4,8))

#Calculate probability densities
alpha = 1.66
amin = 10
x = np.linspace(min(areas), max(areas), 5000)
cpd = (amin / x)**(alpha - 1)
dataCPD = normed_CPD(data['area'])

#Plot probability densities #dodgerblue '#4d9fcc'
ax[0].loglog(np.sort(areas), dataCPD, linewidth = 3, color = c2)#, markerfacecolor = 'none')
ax[0].loglog(x, 1.4*cpd, linestyle = '-', linewidth = 1, label = '$\\overline{F}(A) \propto A^{-0.675} $', color = 'k')
ax[0].legend(loc = (.35, .75), fontsize = 12, framealpha = 1, handlelength = 0, frameon = False)
ax[0].set_xlim([10**0.8, 10**5.6])
ax[0].set_ylabel('$\\overline{F}(A)$', fontsize = 10)
ax[0].get_xaxis().set_ticks([10**1, 10**3, 10**5])
ax[0].get_yaxis().set_ticks([10**-4, 10**-3, 10**-2, 10**-1, 10**0])
ax[0].tick_params('both', length = 10, width = 1.2, which = 'major', labelsize = 8)
ax[0].tick_params('both', length = 0, width = 1, which = 'minor')
ax[0].annotate('(a)', xy = (-.3,1.1), xycoords = 'axes fraction', fontsize = 12)

#Calculate the fractal perimeter dimension
sqAreas = np.sqrt(areas)
D, b, r, p, se = linregress(np.log10(np.sqrt(areas)), np.log10(perimeters))

#Plot fractal dimensions #forestgreen
ax[1].loglog(areas_without, perimeters_without, 'v', markersize = 5, markerfacecolor = c4, markeredgecolor = 'w', markeredgewidth = 0.2, label = 'no ponds')
ax[1].loglog(areas_with, perimeters_with, 'o', markersize = 4, markerfacecolor = "#000000", markeredgecolor = 'w', markeredgewidth = 0.4, label = 'ponded')
ax[1].loglog(x, 10**b*x**(D/2), linestyle = '-', linewidth = 1, color = 'k')
ax[1].set_xlim([10**0.8, 10**5.6])
ax[1].set_ylabel('Perimeter [m]', fontsize = 10)
ax[1].get_xaxis().set_ticks([10**1, 10**3, 10**5])
ax[1].tick_params('both', length = 10, width = 1.2, which = 'major', labelsize = 8)
ax[1].tick_params('both', length = 0, width = 1, which = 'minor')
ax[1].annotate('$P \propto A^{0.58}$', xy = (.23, .6), xycoords = 'axes fraction', fontsize = 12)
ax[1].annotate('(b)', xy = (-0.3,1.1), xycoords = 'axes fraction', fontsize = 12)
ax[1].legend(loc = 'lower right', fontsize = 10)

#Calculate depth-area scaling exponent
d, yint, rval, pval, se = linregress(np.log10(areas), np.log10(depths))

#Plot depth-area scaling
ax[2].loglog(areas_without, depths_without, 'v', markersize = 5, markerfacecolor = c5, markeredgecolor = 'w', markeredgewidth = 0.2, label = 'no ponds')
ax[2].loglog(areas_with, depths_with, 'o', markersize = 4, markerfacecolor = '#000000', markeredgecolor = 'w', markeredgewidth = 0.4, label = 'ponded')
ax[2].loglog(x, 10**yint*x**d, '-', c = 'k', linewidth = 1) 
ax[2].get_xaxis().set_ticks([10**1,10**3,10**5])
ax[2].set_ylabel('Depth [m]', fontsize = 10)
ax[2].tick_params(labelsize = 8)
ax[2].tick_params('both', length = 10, width = 1.2, which = 'major')
ax[2].annotate('$d \propto A^{0.49}$', xy = (0.1,0.77), xycoords = 'axes fraction', fontsize = 12)
ax[2].set_xlabel('Depression Area [m$^2$]', fontsize = 10)
ax[2].annotate('(c)', xy = (-.3,1.1), xycoords = 'axes fraction', fontsize = 12)
ax[2].legend(loc = 'lower right', fontsize = 10)

plt.minorticks_off()
sns.despine(offset = 5)
fig.subplots_adjust(hspace = 0.15, wspace = 0.25)
plt.tight_layout()

fig.set_facecolor('white')

plt.savefig('figure-2_GRL.png', dpi = 450)
plt.show()
```

```python
dataCPD1 = normed_CPD(areasList1)
dataCPD2 = normed_CPD(areasList2)
dataCPD3 = normed_CPD(areasList3)
dataCPD4 = normed_CPD(areasList4)

dataCPD5 = normed_CPD(areasList5)
dataCPD6 = normed_CPD(areasList6)
dataCPD7 = normed_CPD(areasList7)
dataCPD8 = normed_CPD(areasList8)
```
